```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("../data/nonca_kmers.csv")
test=df[df['count'] != 1]
df
```

```python
df=df.groupby(['count']).count()
```

```python
df
```

```python
ax=sns.lineplot(data=df, x="count",y="number")
ax.set_ylabel("Frequency")
ax.set_xlabel("Kmer Counts")
sns.despine()
plt.savefig('line.png',bbox_inches ="tight",dpi=300)
```

```python
sns.set_context("paper")
sns.color_palette("pastel")
sns.set(rc={'figure.figsize':(10,10)})
sns.set_style("white")
ax=sns.histplot(df,x="count",kde=True)
ax.set_ylabel("Frequency")
ax.set_xlabel("Kmer Counts")
sns.despine()
plt.savefig('hist.png',bbox_inches ="tight",dpi=300)
```

```python
sns.set_context("paper")
sns.color_palette("pastel")
sns.set(rc={'figure.figsize':(10,10)})
sns.set_style("white")
ax=sns.histplot(test,kde=True)
ax.set_ylabel("Count")
ax.set_xlabel("Frequency of Counts")
sns.despine()
```
